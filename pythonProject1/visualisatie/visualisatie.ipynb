## __Plot1__
Een duidelijke plot, met goede markeringen, en hoge contrast kleuren.
dit maakt het makkelijk af te lezen, en zo kan je dus duidelijk de verschillen zien tussen de voedingsmiddelen.

```python
import matplotlib.pyplot as plt

dataset = {
    "macadamia": 718,
    "boter": 716,
    "pindakaas": 588,
    "pecannoten": 690,
    "zonnebloempitten": 584,
    "salami": 507,
    "popcorn": 374,
    "aardappel": 88,

}

food_names = []
calories_amount = []
for foods in dataset:
    food_names.append(foods)
    calories_amount.append(dataset[foods])

fig = plt.figure()
ax = fig.add_axes([0,0,1.5,1])
ax.bar(food_names, calories_amount, zorder = 3)
fig.patch.set_facecolor('darkgrey')
ax.set_facecolor("lightgray")
ax.grid(zorder=0, color=("black"))
plt.ylabel("<----------CalorieÃ«n per 100g------------->")
plt.xlabel("<----------Het voedingsmiddel------------->")
plt.show()
```

## __plot2__
Een slechte plot, lage contrast kleuren, verwarrende achtergrond, en geen duidelijke markering van hoeveelheid.
waardoor het zeer slecht af te lezen is

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

dataset = {
    "macadamia": 718,
    "boter": 716,
    "pindakaas": 588,
    "pecannoten": 690,
    "zonnebloempitten": 584,
    "salami": 507,
    "popcorn": 374,
    "aardappel": 88,
}

food_names = []
calories_amount = []

for foods in dataset:
    food_names.append(foods)
    calories_amount.append(dataset[foods])

# if you enter an invalid rgba value, it will render a stinkbug :)
colors = ['#FFFF00', '#FFFF52', '#FFFF42', '#A50000']

fig, ax = plt.subplots()
img = mpimg.imread('./stinkbuggy.png')
ax.imshow(img, extent=[-20, 100, -5, 45])
ax.pie(calories_amount, labels=food_names, colors=colors, startangle=10)
fig.patch.set_facecolor('midnightblue')
plt.show()
```
